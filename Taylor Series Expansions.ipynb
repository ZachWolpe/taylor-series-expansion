# Taylor Series Formula:

A function $f(x)$ is approximated by a polynomial, expanded in a specific point, $x = a$. The general
expression for a Taylor series expansion of the function $f(x)$ in $a$ is

$$f(x) = f^1(a)(x-a) + \frac{f^2}{2!}(x-a)^2 + \frac{f^3}{3!}(x-a)^3 ... + \frac{f^n}{n!}(x-a)^n + R_n$$


$$ = \sum_{n=1}^\infty \frac{f^n(a)}{n!}(x-a)^n + R_n$$


where:
 - $R_n$ is the remainder if a polynomial of degree $n$ is used
 - $f^n(a)$ the $n^{th}$ derivative of $f(x)$ evaluated in the point $x = a$
 - $f^0(a) = f(a)$ the function value of $f(x)$ in the point $x = a$
