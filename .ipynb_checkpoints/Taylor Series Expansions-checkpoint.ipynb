Taylor Series Expansions are a method - used frequently in calculus and physics - to approximate complex function with __POLYNOMIALS__.

__POLYNOMIALS__ are easily differentialable, integratable simpler, more computationally sound, and thus easier to work with.

# Taylor Series Formula:

A function $f(x)$ is approximated by a polynomial, expanded in a specific point, $x = a$. The general
expression for a Taylor series expansion of the function $f(x)$ in $a$ is

$$f(x) = f^1(a)(x-a) + \frac{f^2}{2!}(x-a)^2 + \frac{f^3}{3!}(x-a)^3 ... + \frac{f^n}{n!}(x-a)^n + R_n$$


$$ = \sum_{n=1}^\infty \frac{f^n(a)}{n!}(x-a)^n + R_n$$


where:
 - $R_n$ is the remainder if a polynomial of degree $n$ is used
 - $f^n(a)$ the $n^{th}$ derivative of $f(x)$ evaluated in the point $x = a$
 - $f^0(a) = f(a)$ the function value of $f(x)$ in the point $x = a$

### Taylor Expansion Implementation

The function below allows use to pass in any function and the required polynomial expansion and returns the Taylor series computed function

#### steps
 - for each point in $x$
  - for each polynomial term
   - compute the value of $y$ and sum to current $x$

````python
import numpy as np
import pandas as pd
from scipy.misc import derivative as df
import math
import matplotlib.pyplot as plt
#from matplotlib.pyplot import interact
plt.style.use('seaborn')
````

````python
def taylor_series_expansion(function, a, p, x):
    '''
    Return:
        polynomial functional form of computed by taylor series expansion
        
    Parameters:
        function: the function to utilize 
        a: point at which to evaluate function 
        p: the degree polynomial to compute
        x: the points at which to evaluate the function
    '''
    
    n = len(x)
    
    expansion_y = np.array([])
    for i in x:
        # for each x, to return y values
        fa = 0
        
        for j in range(1,p+1):
        # for each polynomial degree
            fa = fa + (df(function, a, n=j) / math.factorial(n)) * (i - a)**n
            
        np.append(expansion_y, fa)   

        
    return expansion
````
